```python
import pandas as pd
```

```python
# 导入
dt = pd.read_csv('assignment-01-data-unformated.csv', header=1)
```

```python
dt.head(10)
```

#### 修改数据

```python
# 生成地点名列
ls_site = dt['name'].tolist()
ls_site
```

```python
sites = []
for i in range(1, 41):
    sites += ls_site
    i+=1
    
print(len(sites))

# 新的name列
col_site = pd.Series(sites)
```

```python
# 生成经度
long = dt['longitude'].tolist()
longis= []
for i in range(1, 41):
    longis += long
    i+=1
    
print(len(longis))

# 新的经度列
col_long = pd.Series(longis)
```

```python
# 生成纬度
lats = dt['latitude'].tolist()
latis= []
for i in range(1, 41):
    latis += lats
    i+=1
    
print(len(latis))

# 新的纬度列
col_lat = pd.Series(latis)
```

```python
# 珊瑚种类
coral_ls = ["soft corals"]*64 + ["sea fans"]*64 + ["blue corals"]*64 + ["hard corals"]*64 + ["sea pens"]*64
print(len(coral_ls))

# 珊瑚种类列
col_type = pd.Series(coral_ls)
```

```python
# 拆分白化列
ls_cols = dt.columns.tolist()
print(len(ls_cols))
ls_cols2 = ls_cols[3:]
```

```python
dt2 = dt.loc[:, ls_cols2]
dt2.head()
```

```python
# 白化率宽数据变长数据
sub_bleach = pd.melt(dt2, value_vars=ls_cols2, var_name='Year', value_name='bleaching')

sub_bleach.shape[0]
```

```python
sub_bleach.head(80)
```

```python
#新建年份列，用于合并
y1 = []
sub = 2017

for i in range(1,9):
    y1 += [sub]*8
    sub-=1

y2 = []

for j in range(1,6):
    y2 += y1
    
# 生成年份列
col_year = pd.Series(y2)
```

```python
# 提取白化率
col_bleach = sub_bleach.loc[:, 'bleaching']
```

#### 整合成新表

```python
new_dict = {'name':col_site,'longitude':col_long,'latitude':col_lat,'Coral Type':col_type, 'Year': col_year, 'Bleaching': col_bleach}
new_set = pd.DataFrame(new_dict)
```

```python
output = new_set.to_csv('wrangled_bleach.csv', index=False)
```
